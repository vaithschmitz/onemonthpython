# Import Pandas

```python
import pandas as pd
```

# Importing Data

#### names is header row in pd, sep == separator (t == tab)), index_col is col to be used for index

```python
names = ['id', 'title', 'year', 'rating', 'votes', 'length', 'genres']
data = pd.read_csv('imdb_top_10000.txt', sep='\t', names = names, index_col=0)
```

#### Number of elements in columns + types of elements

```python
data.info()
```

#### More info about data in numeric columns => count, mean, standard deviation, quartiles, min, max

```python
data.describe()
```

# Exporting Data

```python
data.to_csv('test.csv', header=True, index=True, sep=',')
```

# Plotting Data

#### Running inline to show plots on Jupyter frontend

```python
%matplotlib inline
```

```python
data.plot()
```

#### Plot types 

```python
data.plot(kind='scatter', x='year', y='rating')
```

#### set alpha for better density viz (docs for this at pandas plot)

```python
data.plot(kind='scatter', x='year', y='rating', alpha=0.3)
```

#### Plot single column by frequency

```python
data.rating.plot(kind='hist')
```

```python
import seaborn as sns
```

#### seaborn automatically calculate regression line

```python
sns.lmplot(x='year', y='rating', data=data, scatter_kws={'alpha':0.3})
```

#### Pairplot automatically plots all numeric columns against each other

```python
sns.pairplot(data)
```

# Ordinary Least Squares (OLS) Regression 

```python
import statsmodels.api as sm
```

```python
results = sm.OLS(data['year'], data['rating']).fit()
```

```python
results.summary()
```
